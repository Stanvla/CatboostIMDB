```python
import catboost as cb
import os
import pandas as pd
import numpy as np
```

```python
dummy_x = np.random.random((10, 3))
dummy_y = (np.random.random(10) > 0.5) + 0
```

```python
dummy_x, dummy_y
```

```python
df = pd.DataFrame(data=dummy_y, columns=['labels'])
```

```python
df['emb'] = dummy_x.tolist()
df['num'] = np.random.random(10)
```

```python
df
```

```python
pool = cb.Pool(
    data=df.drop(labels=['labels'],axis=1),
    label=df.labels,
    embedding_features=['emb']
)
```

```python
classifier = cb.CatBoostClassifier()
```

```python
classifier.fit(pool)
```

```python
regressor = cb.CatBoostRegressor()
```

```python
regressor.fit(pool)
```
